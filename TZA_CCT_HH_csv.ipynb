```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming the uploaded file is named 'TZA_CCT_HH.csv'
df = pd.read_csv('TZA_CCT_HH.csv')

# 1. Histogram of a numerical column (e.g., 'hh_size'):
plt.figure(figsize=(8, 6))
sns.histplot(df['drought_flood'], bins=20, kde=True)
plt.title('Distribution of Household Size')
plt.xlabel('drought_flood')
plt.ylabel('Frequency')
plt.show()



```
