```python
import collections

def find_most_frequent_words(filename, top_n=10):
    with open(filename, 'r') as file:
        text = file.read()

    words = text.split()
    word_counts = collections.Counter(words)
    most_frequent = word_counts.most_common(top_n)
    return most_frequent

# Example usage:
obama_most_frequent = find_most_frequent_words("obama_speech.txt")
print(obama_most_frequent)
```

```python
import csv

def count_lines_in_csv(filename, search_terms):
    with open(filename, 'r') as csvfile:
        reader = csv.reader(csvfile)
        count = 0
        for row in reader:
            for term in search_terms:
                if term in row[0]:  # Assuming search in the first column
                    count += 1
                    break  # Move to the next row if a match is found
    return count

# Example usage:
python_count = count_lines_in_csv("hacker_news.csv", ["python", "Python"])
print(python_count)
```
